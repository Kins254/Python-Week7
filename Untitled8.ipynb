People's musical tastes change as they get older (my parents, for example, after years of listening to relatively cool music when I was a kid, subsequently hit their mid-fifties and developed a worrying obsession with country and western music). So, I did some  research to find out if it's possible to be old and maintain music preference too. First, I got two groups of people (45 people in each group): one group contained young people (which I arbitrarily decided was under 40 years of age) **( encoded as 1)** and the other group contained more mature individuals (above 40 years of age) **(encoded as 2)**. This is my first independent variable, age. I then split each of these groups of 45 into three smaller groups of 15 and assigned them to listen to Band-X (who everyone knows are the coolest band on the planet) **(encoded as 1)** , ABBA **(encoded as 2)** ,then  Barf Grooks (a less well-known country and western musician) **(encoded as 3)**. This is my second independent variable, music. After listening to the music I got each person to rate it on a scale ranging from -100 (please poke a pencil through my eardrum so I don't have to listen any more) through 0 (I am completely indifferent) to +100 (I love this music so much, it gives me a tingle down my spine). This variable is called liking. The data are in the file music.csv.

### Grouping And Visualization

```python
age_distribution = music['AGE'].value_counts()
music_distribution = music["MUSIC"].value_counts()

print("age distribution \n",age_distribution)
print("Music distribution \n", music_distribution)
```

```python
# Convert AGE to categorical and label
music['AGE'] = pd.Categorical(music['AGE'], categories=[1, 2], ordered=True)
music['AGE'] = music['AGE'].cat.rename_categories({
    1: 'Below 40yrs',
    2: 'Above 40yrs'
})

print(music['AGE'])

# Convert MUSIC to categorical and label
music['MUSIC'] =pd.Categorical(music['MUSIC'],categories=[1,2,3],ordered=True)
music['MUSIC']= music['MUSIC'].cat.rename_categories({
    1: 'Band-X',
    2: 'ABBA',
    3: 'Barf Grooks'
})
print(music['MUSIC'])

#Grouping likes into bins from -150 to 150
# Defining the bin edges and labels
bins = [-150, -50, 50, 100, 150]
labels = ['Very Negative', 'Negative', 'Positive', 'Very Positive']

# Creating a new column 'LIKING_BIN' with the categorized data
music['LIKING_BIN'] = pd.cut(music['LIKING'], bins=bins, labels=labels, right=True)
liking_bin_dist=music['LIKING_BIN'].value_counts()
print('liking bin counts:',liking_bin_dist)

# Display the first few rows to see the results
print(music[['LIKING', 'LIKING_BIN']].head())

```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the histogram for the 'LIKING' variable to know if it is normally distributed or not
# Spliting the data into positive and negative values
positive_liking = music[music['LIKING'] > 0]['LIKING']
negative_liking = music[music['LIKING'] < 0]['LIKING']

# Creating a figure with two subplots
plt.figure(figsize=(12, 6))

# Ploting for positive liking values
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.histplot(positive_liking, bins=20, kde=True, color='skyblue')
plt.title('Positive Liking Distribution')
plt.xlabel('Liking')
plt.ylabel('Frequency')

# Ploting for negative liking values
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.histplot(negative_liking, bins=20, kde=True, color='lightcoral')
plt.title('Negative Liking Distribution')
plt.xlabel('Liking')
plt.ylabel('Frequency')

# Adjusting layout and showing the plots
plt.tight_layout()
plt.show()
```

The two graph appears to follow a normal distribution. Let's confirm this by using the Shapiro-Wilk test

### Shapiro test for normlity

```python


# Shapiro-Wilk test for normality on the 'LIKING' column (positive values only)
positive_liking = music[music['LIKING'] > 0]['LIKING']
stat, p_value = stats.shapiro(positive_liking)

print('Shapiro-Wilk Test for Positive Liking:')
print('Statistic:', stat)
print('p-value:', p_value)

# Shapiro-Wilk test for normality on the 'LIKING' column (negative values only)
negative_liking = music[music['LIKING'] < 0]['LIKING']
stat, p_value = stats.shapiro(negative_liking)

print('\nShapiro-Wilk Test for Negative Liking:')
print('Statistic:', stat)
print('p-value:', p_value)

# Interpretation
if p_value > 0.05:
    print("\nThe data follows a normal distribution (fail to reject H0).")
else:
    print("\nThe data does not follow a normal distribution (reject H0).")
```

### Line graph for liking and music 

```python
# Line chart for average liking across music genres
avg_liking = music.groupby('MUSIC')['LIKING'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.lineplot(data=avg_liking, x='MUSIC', y='LIKING', marker='o', color='blue')
plt.title('Average Liking Score Across Music Genres')
plt.xlabel('Music Genre')
plt.ylabel('Average Liking Score')
plt.show()
```

From the line graph above, we can see that the most liked music genre is ABBA, followed by Barf Grooks, and lastly, Band-X.

### Heatmap for correlation

```python
# Heatmap for correlation
correlation_matrix = music[['LIKING', 'AGE','MUSIC']].apply(lambda col: col.cat.codes if col.dtype.name == 'category' else col).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

From the above correlation heatmap, we can observe that the correlation between music and liking is a positive 0.04. This suggests that a unit increase in the music type (e.g., moving from Band-X to ABBA) is associated with a 0.04 increase in the liking score on average. Additionally, there is no correlation between music and age, as the correlation value is 0.00.
